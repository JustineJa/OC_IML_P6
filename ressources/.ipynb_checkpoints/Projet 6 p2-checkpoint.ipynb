# Objectif

L'objectif de cette deuxième partie est d'utiliser des techniques plus avancées de NLP et de reconnaissance d'images pour classifier nos observations. L'idéal serait d'obtenir une bijection entre la classification actuelle et la nouvelle, c'est-à-dire un score ARI élevé.

# Chargement des packages

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.lines import Line2D
from mpl_toolkits import mplot3d
import pandas as pd
import seaborn as sns
import os
import missingno as msno
import warnings
import time
warnings.simplefilter(action='ignore', category=FutureWarning)
from sklearn.metrics import silhouette_samples, silhouette_score, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn import preprocessing
from sklearn.model_selection import GroupShuffleSplit, train_test_split
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import keras
from tensorflow.compat.v1.keras import backend as KB
from keras.preprocessing.image import load_img, img_to_array
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.layers import Dense, Flatten, Input, Lambda
from keras.models import Model, load_model
from keras import optimizers
from keras.callbacks import EarlyStopping
import tensorflow.compat.v1 as tf
# On désactive la eager execution pour rendre tensorflow compatible avec nos scripts
tf.disable_eager_execution()
import tensorflow_hub as hub
import pickle
import umap
```

# Chargement du dataset

```python
data = pd.read_csv("./data/clean_data.csv")
```

# Approche n°3 : Reconnaissance d'images avancée

Le dataset d'étude contient peu d'observations, tandis que la reconnaissance d'images est un procédé faisant intervenir un grand nombre de dimensions et de paramètres. Par conséquent, nous avons un terrain propice à la sur-interprétation, qui rendrait un modèle assez peu performant, comme on a pu l'observer lors de l'approche n°2.

Pour contrer cela, nous allons utiliser un réseau de neurones pré-entraîné sur un grand nombre d'images, que nous allons adapter pour notré étude via un transfer learning. 

Nous utiliserons ici le réseau VGG-16

## Chargement des images

La prmière étape consistera à charger toutes les images en les pré-traitant pour qu'elles soient compatibles avec un réseau VGG-16

```python
img_dict = {}
start_clock = time.time()
for item_id in data["uniq_id"]:
    item_image_id = data.loc[data["uniq_id"] == item_id, "image"].iloc[0]
    item_image_path = "./data/Flipkart/Images/" + item_image_id
    item_image = load_img(item_image_path, target_size=(224, 224))
    item_image = img_to_array(item_image)
    img_dict[item_id] = item_image    
print("Images chargées en {}s".format(round(time.time() - start_clock)))
```

```python
img_matrix = np.array([array2d for array2d in img_dict.values()])
```

```python
img_matrix.shape
```

```python
img_matrix = preprocess_input(img_matrix)
```

## Chargement du réseau de neurones

```python
# Chargement du VGG-16 pré-entraîné sur ImageNet et sans les couches fully-connected
model_vgg16 = VGG16(weights = "imagenet", include_top = False, input_shape = (224, 224, 3))

# Récupération de la sortie du réseau
otp = model_vgg16.output

# Ajout de la nouvelle couche fully-connected pour la classification à 7 classes
otp = Flatten(name = "flatten")(otp)
new_predictions = Dense(7, activation = 'softmax')(otp)

# Création du nouveau modèle
new_model = Model(inputs = model_vgg16.input, outputs = new_predictions)
```

## Séparation train-test

```python
lb = preprocessing.LabelBinarizer()
```

```python
y = data["main_category"].to_numpy()
lb.fit(y)
X_train, X_test, y_train, y_test = train_test_split(img_matrix, y, test_size=0.3, random_state=42)
y_train = lb.transform(y_train)
X_train = preprocess_input(X_train)
X_test = preprocess_input(X_test)
```

## Entraînement par feature extraction supervisé

### Entraînement

Etant donné qu'on applique du transfer learning, on indique au modèle que seule la couche Dense est à entraîner.

```python
for layer in new_model.layers[:-1]:
    layer.trainable = False
```

Notre modèle a donc la structure suivante :

```python
new_model.summary()
```

Nous allons maintenant entraîner notre réseau de neurones. Nous lui demanderons de contrôler l'exactitude de ses prédictions sur un sous-ensemble de validation croisés et de ne retenir que le meilleur score d'exactitude sur ce jeu. Nous lui indiquerons également qu'il doit arrêter l'entraînement quand il ne progresse plus, même s'il n'a pas réalisé toutes ses époques.

```python
new_model.compile(loss='categorical_crossentropy', optimizer = optimizers.SGD(lr = 0.0001, momentum = 0.9), 
                  metrics=["accuracy"])
callback = EarlyStopping(monitor='val_loss', patience=3, min_delta = 0.005, restore_best_weights = True)
```

```python
if not os.path.exists("./data/new_model"):
    start_clock = time.time()
    new_model.fit(X_train, y_train, epochs = 30, batch_size = 100, verbose=2, validation_split = 0.2, 
                  callbacks = [callback])
    print("Modèle entraîné en {}s".format(round(time.time() - start_clock)))
elif os.path.exists("./data/new_model"):
    load_model('./data/new_model')
```

Sans avoir été particulièrement long, cet entraînement n'a pas pour autant été court ; on choisira donc d'enregistrer le modèle pour le réutiliser plus facilement si l'on souhaite rouvrir le notebook à l'avenir.

```python
if not os.path.exists("./data/new_model"):
    new_model.save("./data/new_model")
    print("modèle Keras enregistré")
```

On obtient un score assez élevé sur le sous-ensemble de validation du jeu d'entraînement. regardons comment le modèle extrapole pour le jeu de test

### Test

```python
model_prediction = new_model.predict(X_test)
```

```python
labeled_prediction = lb.inverse_transform(model_prediction)
```

```python
accuracy_score(y_true = y_test, y_pred = labeled_prediction)
```

```python
supervized_model_score = {}
supervized_model_score["Reconnaissance d'images"] = accuracy_score(y_true = y_test, y_pred = labeled_prediction)
```

Le score d'exactitude est également très élevé sur le jeu de test.
Regardons où sont les erreurs de notre modèle

```python
cm = confusion_matrix(y_true = y_test, y_pred = labeled_prediction)
```

```python
fig, ax = plt.subplots(figsize = (18, 8))
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = data["main_category"].unique())
disp.plot(ax = ax)
plt.xticks(rotation = 45, rotation_mode = "anchor", ha = "right")
```

On remarque que les erreurs sont majoritairement dues à une confusion entre les catégories *Home Furnishing*, *Home Decor* et *Kitchen & Dining*, qui sont intuitivement trois catégories assez proches dans leur nature. Cette confusion pourrait être corrigée avec plus de données d'entraînement.

On remarque néanmoins que le modèle parvient à faire une classification très précise malgré le fait qu'on ait très peu de données. Cette approche est donc très prometteuse : on peut entraîner ce modèle sur un peu plus de données puis le laisser classer les futures données qui seront ajoutées par la suite dans le catalogue.

Cependant, cette approche consiste en un apprentissage supervisé et s'écarte, malgré son efficacité, de l'objectif d'origine de l'étude qui reposait principalement sur une classification non supervisés.

Pour cela, nous allons reprendre notre réseau de neurones en lui supprimant sa couche Dense.

## Extraction de features pour apprentissage non supervisé

### Modèle

On recharge le modèle, cette fois sans ajouter de couche Dense.

```python
# Chargement du VGG-16 pré-entraîné sur ImageNet et sans les couches fully-connected
model_vgg16 = VGG16(weights = "imagenet", include_top = False, input_shape = (224, 224, 3))

# Récupération de la sortie du réseau
otp = model_vgg16.output

# Ajout de la couche Flatten
new_predictions = Flatten(name = "flatten")(otp)

# Création du nouveau modèle
fe_model = Model(inputs = model_vgg16.input, outputs = new_predictions)
```

On transforme directement notre matrice d'images en features grâce à ce modèle :

```python
vgg_pred_array = fe_model.predict(img_matrix)
```

### Clustering

Comme dans le notebook 1, nous essaierons le clustering avec 11 composantes principales et avec 80% de variance expliquée.

```python
pca_80 = pca.transform(vgg_pred_array)
```

```python
pca_11 = PCA(11, random_state = 42).fit_transform(vgg_pred_array)
```

```python
pca_80.shape
```

```python
pca_11.shape
```

#### 80% de variance expliquée

```python
inertias = []

K = range(2, 21)

start_clock = time.time()

for k in K:
    print("Evaluation pour k = {}...".format(k))

    # Création du modèle
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(pca_80)
    
    # Calcul de l'intertie
    inertias.append(kmeanModel.inertia_)

print("Evaluation des différentes méthodes en {}s".format(round(time.time() - start_clock)))

fig, ax = plt.subplots(figsize = (18, 8))
ax.plot(K, inertias, 'bx-')
ax.set_xlabel('Valeurs de K', fontsize = 14)
ax.set_ylabel('Inertie', fontsize = 14)
ax.set_xticks([i for i in range(21)])
plt.suptitle('Elbow Method par inertie', fontsize = 18)
plt.show()
```

```python
silh_score_dict = {}
for n_clusters in K:

    clusterer = KMeans(n_clusters = n_clusters, random_state = 42)
    cluster_labels = clusterer.fit_predict(pca_80)
    silhouette_avg = silhouette_score(pca_80, cluster_labels)
    silh_score_dict[n_clusters] = silhouette_avg
    print("Pour n_clusters =", n_clusters,
          "Le score de silhouette moyen est de :", silhouette_avg)

fig, ax = plt.subplots(figsize = (18, 8))
ax.bar(silh_score_dict.keys(), silh_score_dict.values())
ax.set_xticks(K);
plt.suptitle("Score de silhouette en fonction du K-means appliqué", fontsize = 18, fontweight = "bold");
ax.set_xlabel("Valeur de K", fontsize = 16);
ax.set_ylabel("Score de silhouette", fontsize = 16);
max_value = max(silh_score_dict.values())
ax.axhline(max_value, linestyle = "--", color = "red")
ax.text(np.array(K).mean(), 0.97*max_value, str("max = " + str(round(max_value, 2))), color = "red", fontsize = 14, 
        fontweight = "bold");
```

Les scores de silhouettes sont très faibles. Nous n'irons pas plus loin dans cette méthode.

#### 11 composantes principales

```python
inertias = []

K = range(2, 21)

start_clock = time.time()

for k in K:
    print("Evaluation pour k = {}...".format(k))

    # Création du modèle
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(pca_11)
    
    # Calcul de l'intertie
    inertias.append(kmeanModel.inertia_)

print("Evaluation des différentes méthodes en {}s".format(round(time.time() - start_clock)))

fig, ax = plt.subplots(figsize = (18, 8))
ax.plot(K, inertias, 'bx-')
ax.set_xlabel('Valeurs de K', fontsize = 14)
ax.set_ylabel('Inertie', fontsize = 14)
ax.set_xticks([i for i in range(21)])
plt.suptitle('Elbow Method par inertie', fontsize = 18)
plt.show()
```

L'analyse elbow donne une grande coudée pour une valeur de K aux alentours de 8.

```python
silh_score_dict = {}
for n_clusters in K:

    clusterer = KMeans(n_clusters = n_clusters, random_state = 42)
    cluster_labels = clusterer.fit_predict(pca_11)
    silhouette_avg = silhouette_score(pca_11, cluster_labels)
    silh_score_dict[n_clusters] = silhouette_avg
    print("Pour n_clusters =", n_clusters,
          "Le score de silhouette moyen est de :", silhouette_avg)

fig, ax = plt.subplots(figsize = (18, 8))
ax.bar(silh_score_dict.keys(), silh_score_dict.values())
ax.set_xticks(K);
plt.suptitle("Score de silhouette en fonction du K-means appliqué", fontsize = 18, fontweight = "bold");
ax.set_xlabel("Valeur de K", fontsize = 16);
ax.set_ylabel("Score de silhouette", fontsize = 16);
max_value = max(silh_score_dict.values())
ax.axhline(max_value, linestyle = "--", color = "red")
ax.text(np.array(K).mean(), 0.97*max_value, str("max = " + str(round(max_value, 2))), color = "red", fontsize = 14, 
        fontweight = "bold");
```

K = 2 semble être un nombre assez faible de clusters. On retiendra donc K = 4

```python
for K in [4]:
    kmeans = KMeans(n_clusters = K, random_state = 42)
    kmeans.fit(pca_11)
    kmeans_labels = kmeans.predict(pca_11)

    km_labs_distribution = pd.DataFrame(np.array(np.unique(kmeans_labels, return_counts = True)).T)
    km_labs_distribution.rename(columns = {0 : "cluster", 1 : "count"}, inplace = True)
    km_labs_distribution["proportion"] = km_labs_distribution["count"].apply(lambda X : round(100 * X / km_labs_distribution["count"].sum()))

    fig, ax = plt.subplots(figsize = (18, 8))
    ax.bar(km_labs_distribution["cluster"], km_labs_distribution["proportion"])
    #ax.set_xticks(dbs_labs_distribution["cluster"]);
    ax.set_xlabel("Cluster", fontsize = 14)
    ax.set_ylabel("% de points inclus dans le cluster", fontsize = 14)
    plt.suptitle("Distribution des points par cluster pour K-means, K = {}".format(K),
                 fontsize = 18)
```

## Création du word embedding

## Modèle supervisé

Nous allons ici déployer un modèle de classification supervisé.

Nous allons pour cela construire une pipeline automatisée qui fonctionne selon les étapes suivantes :

1. Définition des données X et de la sorte y
2. Encodage de X et y
3. Création d'un modèle
4. Entraînement
5. Test

Ici , nos X et y sont clairs : X est l'ensemble des descritpions avant transformation, c'est à dire "messages", et y est l'ensembe des catégories associées

```python
X = messages
y = data["main_category"].to_numpy()
```

Tout comme pour la détection d'images, on pré-process y pour en faire une array de dimension 7. Pour cela, on crée des fonctions d'encodage et de décodage à partir du LabelEncoder de scikit-learn.preprocessing :

```python
le = preprocessing.LabelEncoder()
le.fit(y)
```

```python
def encode(le, labels):
    enc = le.transform(labels)
    return keras.utils.to_categorical(enc)

def decode(le, one_hot):
    dec = np.argmax(one_hot, axis = 1)
    return(le.inverse_transform(dec))
```

Nos fonctions fonctionnent donc ainis pour l'encodage :

```python
test = encode(le, ["Watches", "Home Furnishing", "Home Furnishing", "Baby Care"])
test
```

Et le décodage :

```python
untest = decode(le, test)
untest
```

On encode donc X et y (à vrai dire X sera encodée plus tard par l'universal sentence encoding), et on procède à un train-test split

```python
X_enc = X
y_enc = encode(le, y)

X_train, X_test, y_train, y_test = train_test_split(X_enc, y_enc, test_size=0.3, random_state=42)
```

On obtient des matrices de la forme suivante :

```python
for subelement in (X_train, X_test, y_train, y_test):
    print(subelement.shape)
```

On crée maintenant notre réseau de neurones qui contient :

* Une couche d'input
* Une couche d'embedding, qui sera l'universal sentence encoder
* Une layer Dense qui divisera par deux la dimension
* Un dernière layer Dense de prédiction selon 7 classes

```python
def UniversalSentenceEncoding(x):
    return embed(tf.squeeze(tf.cast(x, tf.string)))
```

```python
input_layer = Input(shape = (1,), dtype = tf.string)
embedding_layer = Lambda(UniversalSentenceEncoding, output_shape = (512,))(input_layer)
dense_layer = Dense(256, activation = "relu")(embedding_layer)
prediction_layer = Dense(7, activation = "softmax")(dense_layer)

nlp_model = Model(inputs = [input_layer], outputs = prediction_layer)
```

On obtient le modèle suivant :

```python
nlp_model.summary()
```

Il contient très peu de paramètres à entraîner car le gros du travail a été fait en amont par google dans son universal sentence encoding. Nous avons donc construit un modèle à la main, mais ce que l'on a fait n'est en réalité rien d'autre qu'un transfer learning.

On peut donc compiler et entraîner le modèle.

```python
nlp_model.compile(loss='categorical_crossentropy', optimizer = "adam", metrics=["accuracy"])
callback = EarlyStopping(monitor='val_loss', patience=3, min_delta = 0.008, restore_best_weights = True)
```

```python
start_clock = time.time()
with tf.Session() as session:
    KB.set_session(session)
    session.run(tf.global_variables_initializer())
    session.run(tf.tables_initializer())
    training_session = nlp_model.fit(X_train, y_train, epochs = 50, batch_size = 50, verbose=0, 
                                     validation_split = 0.3, callbacks = [callback])
    nlp_model.save_weights("./data/nlp_model.h5")
print("Modèle entraîné en {}s".format(round(time.time() - start_clock)))
```

### Exactitude sur le jeu d'entraînement

## Modèle non supervisé

Nous réutiliserons ici notre matrice message_embeddings obtenue dans le paragraphe 5.1.

### Clustering sans réduction dimensionnelle

Nous chercherons comme à notre habitude la valeur optimale de K pour un K-means clustering :

```python
inertias = []

K = range(2, 21)

start_clock = time.time()

for k in K:
    print("Evaluation pour k = {}...".format(k))

    # Création du modèle
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(message_embeddings)
    
    # Calcul de l'intertie
    inertias.append(kmeanModel.inertia_)

print("Evaluation des différentes méthodes en {}s".format(round(time.time() - start_clock)))

fig, ax = plt.subplots(figsize = (18, 8))
ax.plot(K, inertias, 'bx-')
ax.set_xlabel('Valeurs de K', fontsize = 14)
ax.set_ylabel('Inertie', fontsize = 14)
ax.set_xticks([i for i in range(21)])
plt.suptitle('Elbow Method par inertie', fontsize = 18)
plt.show()
```

```python
silh_score_dict = {}
for n_clusters in K:

    clusterer = KMeans(n_clusters = n_clusters, random_state = 42)
    cluster_labels = clusterer.fit_predict(message_embeddings)
    silhouette_avg = silhouette_score(message_embeddings, cluster_labels)
    silh_score_dict[n_clusters] = silhouette_avg
    print("Pour n_clusters =", n_clusters,
          "Le score de silhouette moyen est de :", silhouette_avg)

fig, ax = plt.subplots(figsize = (18, 8))
ax.bar(silh_score_dict.keys(), silh_score_dict.values())
ax.set_xticks(K);
plt.suptitle("Score de silhouette en fonction du K-means appliqué", fontsize = 18, fontweight = "bold");
ax.set_xlabel("Valeur de K", fontsize = 16);
ax.set_ylabel("Score de silhouette", fontsize = 16);
max_value = max(silh_score_dict.values())
ax.axhline(max_value, linestyle = "--", color = "red")
ax.text(np.array(K).mean(), 0.97*max_value, str("max = " + str(round(max_value, 2))), color = "red", fontsize = 14, 
        fontweight = "bold");
```

Nous obtenons des scores de silhouettes très faibles. Nous n'irons pas plus loin dans cette méthode

### Clustering avec réduction dimensionnelle

Comme à notre habitude, nous chercherons à effectuer une PCA qui explique 80% de la variance, et nous étudierons une dimensionnelle avec toutes les composantes principales ainsi obtenues, et une autre avec seulement 11 composantes principales, c'est-à dire 1 composante pour 100 observations du modèle.

```python
pca = PCA(0.8, svd_solver = "full", random_state = 42)
pca.fit(message_embeddings)
pca.n_components_
```

```python
pca_cumsum = pca.explained_variance_ratio_.cumsum()

fig, ax = plt.subplots(figsize = (18, 8))
ax.bar(range(pca_cumsum.shape[0]), pca_cumsum, alpha = 0.7);
ax.set_xlabel("Nombre de composantes principales", fontsize = 14);
ax.set_ylabel("Variance expliquée", fontsize = 14);
ax.set_xticks(range(0, pca_cumsum.shape[0] + 1, 5));

for yvalue in [0.1 * i for i in range(1, 9)]:
    ax.axhline(yvalue, alpha = 0.2, color = "green")
plt.xticks(rotation = 45, rotation_mode = "anchor", ha = "right");
plt.suptitle("Variance expliquée par la PCA en fonction du nombre de composantes principales",
            fontsize = 18, fontweight = "bold");
```

On arrive à conserver beaucoup de variance expliquée avec peu de composantes principales. Ceci est plutôt bon signe. Voyons ce que l'on obtient.

# Conclusion

Nous observons, dans les modèles non supervisés, que les modèles avancés ont un meilleur score de ARI, ce qui signifie que l'on parvient mieux à retrouver les catégories initiales. Cependant, les scores de silhouettes sont nettement moins bons pour ces derniers par rapport aux modèles basiques. 

Les réseaux de neurones étant des "extracteurs de concepts", il n'est pas surprenant que leurs résultats sont difficilement exploitables autrement que par des couches neuronales, et par conséquent que les résultats des algorithmes de classification classiques se dégradent.

L'amélioration du ARI, mais la dégradation du clustering en général. Ces deux phénomènes justifient alors d'étudier un traitement par réseau de neurones complet, au travers d'un apprentissage supervisé.

**En faisant cela, nous observons des résultats excellents, et ce malgré le peu de données que l'on possède.** 

Si l'on souhaite retrouver les catégories initiales, mieux vaut donc privilégier les réseaux de neurones supervisés. Si, en revanche, on souhaite s'affranchir de ces catégories, mieux vaut alors privilégier les méthodes plus basiques, en particulier le NLP par TF-IDF avec un clustering K-Means

**Nous pouvons donc en conclure que notre étude de faisabilité, portée par cette preuve de concept, a été concluante.**

**Nous pourrons donc déployer plus de ressources sur les deux réseaux de neurones, et les entraîner sur des jeux de données plus fournis en phase de développement**. 

A défaut de moyens ou de ressources, on privilégiera le réseau de neurones NLP, qui s'entraîne bien plus rapidement et nécessite moins de paramétrage que celui de reconnaissance d'images, tout en possédant de meilleurs résultats.
